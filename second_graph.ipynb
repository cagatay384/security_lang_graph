```python
from typing import TypedDict, List
from langgraph.graph import StateGraph

class AgentState(TypedDict):
    name : str
    values : List[int]
    result : str

def computer(state: AgentState) -> AgentState:
    state["result"] = f"Hey {state["name"]} your result is:  {sum(state["values"])}"
    return state

graph = StateGraph(AgentState)
graph.add_node("computer", computer)
graph.set_entry_point("computer")
graph.set_finish_point("computer")
app = graph.compile()

answer = app.invoke({"name":"Bob", "values":[1,2,3,4,5]})
answer



```

when we use the invoke we pass some variables into the code/function/method and want that code to perform, and after that the variables(in this case dictionary keys) are assigned with values, performing occurs and the dictionary keys get updated
