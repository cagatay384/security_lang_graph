```python
from typing import TypedDict, Dict
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict): #this is out state schema
    message: str

def greetingNode(state: AgentState)-> AgentState:
    """simple node that adds a greeting message to the state"""
    state["message"] = "Hey " + state["message"] + ", nice to meet you state, I'm greeting you"

    return state
```

```python
graph = StateGraph(AgentState)
graph.add_node("greeter", greetingNode) # we passed two parameters, one -> node name and the other one is the -> functionality/method

graph.set_entry_point("greeter")
graph.set_finish_point("greeter")
```
